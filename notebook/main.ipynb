## Load and normalize data

### add the projet path so jupiter can recognize

```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join("..")))
```

```python
from src import preprocessing

df = preprocessing.load_dataframe('../data/diabetes_data_upload.csv')
df.head()
```

### Transform variables from binary to 0 or 1 value.

```python
df_processed = preprocessing.transform_dataframe(df)
df_processed.head()
```

### Separate Y than X variable and test from train data

```python
x = df_processed.drop(columns=['class'])
y = df_processed['class']

x_train, x_test, y_train, y_test = preprocessing.split_train_test_data(x, y)
```

## Train the Naive Baynes modal and get the predictions with the test data

```python
from src import models
from sklearn.metrics import confusion_matrix, f1_score

nb_model = models.train_naive_bayes(x_train, y_train)

y_pred_nb = nb_model.predict(x_test)

cm = confusion_matrix(y_test, y_pred_nb)
f1 = f1_score(y_test, y_pred_nb)

print("Matriz de Confusão:\n", cm)
print("F1 Score:", f1)
```
